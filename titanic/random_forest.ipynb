### 1. Read and process data

```python
from sklearn.model_selection import train_test_split
```

```python
def read_data(file_name):
    data = pd.read_csv(file_name, index_col=0)
    col = ['Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']
    if file_name == "train.csv":
        col.insert(0, 'Survived')
        data = data.loc[:, col]
        return data.iloc[:, 1:], data.iloc[:, 0]
    else:
        data = data.loc[:, col]
        return data
```

```python
def process_train_data(data):
    data["Name"] = data["Name"].apply(extract_name)
    data["Cabin"] = data["Cabin"].apply(lambda x: "Unknown" if pd.isna(x) else x[0])
    data = pd.get_dummies(data)
    data["Age"].fillna(data["Age"].mean(), inplace=True)
    data["Fare"].fillna(data["Fare"].mean(), inplace=True)
    return data
```

```python
def process_test_data(train_data, test_data):
    test_data = process_train_data(test_data)
    for x in train_data.columns:
        if x not in test_data.columns:
            test_data[x] = 0
    test_data = test_data.loc[:, train_data.columns]
    return test_data
```

```python
X, y = read_data("train.csv")
print(X.shape)
X.head()
```

```python
X = process_train_data(X)
print(X.shape)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### 2. Train model

```python
from sklearn.ensemble import RandomForestClassifier
```

#### 2.1 Random search

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = list(range(1000))
max_features = ['auto', 'sqrt', 'log2']
max_depth = list(range(1, 50))
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf = RandomForestClassifier(random_state=42)
rf_random = RandomizedSearchCV(estimator=rf,
                               param_distributions=random_grid,
                               n_iter=100, 
                               cv=3,
                               verbose=2,
                               random_state=42,
                               n_jobs=-1)
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

### 2.2 Evaluate random search

```python
from sklearn.metrics import roc_auc_score
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    base_model = RandomForestClassifier(random_state=42)
    base_model.fit(X_train, y_train)
    baseline_auc = roc_auc_score(y_test, base_model.predict_proba(X_test)[:,1])
    print(f"baseline's AUC is {baseline_auc}!")
    current_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    print(f"current model's AUC is {current_auc}!")
    assert current_auc > baseline_auc, "Current model's performance is even worse than the baseline!"
```

```python
evaluate_model(rf_random.best_estimator_, X_train, y_train, X_test, y_test)
```

### 2.3 Grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [600, 617, 650],
    'min_samples_split': [8, 10, 12],
    'min_samples_leaf': [2],
    'max_features': ['auto'],
    'max_depth': [45, 47, 50],
    'bootstrap': [True],
}
```

```python
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf,
                           param_grid=param_grid, 
                           cv=3,
                           n_jobs=-1,
                           verbose=2)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
evaluate_model(grid_search.best_estimator_, X_train, y_train, X_test, y_test)
```
